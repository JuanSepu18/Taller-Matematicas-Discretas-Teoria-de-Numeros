```python
#SOLUCIÓN
def expansion(n,b):
  numeros = []
  while n != 0:
    numeros = [n%b] + numeros
    n= n//b

  print(numeros)
```

```python
#EJEMPLOS
expansion(12,2)

expansion(532,10)

expansion(123232,8)
```
