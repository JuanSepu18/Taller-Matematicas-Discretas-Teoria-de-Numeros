```python
import math as math
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def factoriza(n):
  primos = []
  for i in range(1, n+1):
    contador = 0
    for j in range (1, n+1):
      if i%j == 0:
        contador+=1
    if(contador == 2):
      primos.append(i) 
  i = 0
  potencias = []
  numeros = []

  while n!=1:
    contador = 0
    while n%primos[i] == 0:
      n /= primos[i]
      contador+=1
    if(contador != 0):
      potencias.append(contador)
      numeros.append(primos[i])
    i+=1
    
    
  respuestas = np.array([numeros,potencias])
  print(f"[{respuestas[0]},{respuestas[1]}]")
  
      
```

```python
factoriza(60)

factoriza(30)
```

```python
def test(a,b,c,d,e,f):
  factoriza(a)
  factoriza(b)
  factoriza(c)
  factoriza(d)
  factoriza(e)
  factoriza(f)
```

```python
test(25,465,983,651,7,23)
```

```python
lista = np.arange(1,1000)
tiempos = []
for i in lista:
  ti1 = time.time()
  factoriza2(i)
  tf1 = time.time()

  tiempo1 = tf1 - ti1

  ti2 = time.time()
  factoriza2(i)
  tf2 = time.time()

  tiempo2 = tf2 - ti2

  ti3 = time.time()
  factoriza2(i)
  tf3 = time.time()

  tiempo3 = tf3 - ti3

  ti4 = time.time()
  factoriza2(i)
  tf4 = time.time()

  tiempo4 = tf4 - ti4

  ti5 = time.time()
  factoriza2(i)
  tf5 = time.time()

  tiempo5 = tf5 - ti5

  tiempo = (tiempo1+tiempo2+tiempo3+tiempo4+tiempo5)/5
  tiempos.append(tiempo)

plt.plot(lista, tiempos)
plt.legend()
```
