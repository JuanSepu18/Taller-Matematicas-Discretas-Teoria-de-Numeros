```python
def digito_verificacion(n):
  numeros = list(n)
  suma = 0
  for i in range(1,10):
    suma += (int(numeros[i-1]) * i)
    if(suma%11 == 10):
      resultado = "X"
    else:
      resultado = suma%11
    
  return (n + str(resultado) + " Por lo tanto, el digito de verificacion es " + str (resultado))
```

```python
digito_verificacion("123456789")
```
